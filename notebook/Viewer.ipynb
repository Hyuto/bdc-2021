```python
import os
from PIL import Image

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
import math

def show_images(diff):
    row = math.ceil(len(diff) / 5)
    fig, axes = plt.subplots(row, 5, figsize = (20, 5 * row))
    for i in range(row):
        for j in range(5):
            try:
                path = diff[i * 5 + j][0]
                label = diff[i * 5 + j][1]
            except:
                path = None
                label = None
            if path and label != None:
                img = Image.open(path)
                axes[i, j].imshow(img)
                axes[i, j].set_title(f'File : {os.path.basename(path)}\nLabel : {label}', 
                                    fontsize = 14)
                axes[i, j].axis('off')
    fig.tight_layout(pad=0.3)
    plt.show()
```

## Mengecek perbedaan 2 preprocessed planning

```python
dir_ = '../data/preprocessed'
df = pd.read_csv('../data/preprocessed/preprocessed.csv')[["image", "label"]]
```

```python
dari = 1
sampai = 10

data = [(os.path.join(dir_, x[0]), x[1]) for x in df.values][dari-1:sampai]
show_images(data)
```
